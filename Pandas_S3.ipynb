```python
# Renomeando colunas do data frame

df.rename(columns={
    "district":"bairro",
    "rent":"aluguel",
    "garage":"vagas",
    "bedroos":"quartos",
    "address":"endereco",
    "type":"tipo"
}, inplace=True) # inplace significa que irá fazer qualquer substituição que solicitamo no data frame
```

```python
# Realizando filtros de 3 maneiras diferentes:

# Moradias do bairro morumbi
df[df["bairro"] == "Morumbi"]

# Quantas moradias tem no Morumbi?
df[df["bairro"] == "Morumbi"].count()

```

```python
# Moradias do bairro Morumbi e do tipo Apartamento:

df[(df["bairro"] == "Morumbi") & (df["tipo"] == "Apartamento")]
```

```python
# Terceira forma de filtro

# Mostrar as moradias com area maior ou igual a 300 metros quadrados:

df.query("area >= 300 and total <= 6000")
```

```python
# adicionando uma nova coluna ao data frame

df["condominio"] = df["total"] - df["aluguel"]
```
